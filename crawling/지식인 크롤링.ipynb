```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import time
import pandas as pd

chrome_options = Options()
chrome_options.add_argument('--headless') # headless 모드 활성화
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')


# 웹페이지 열기
# query = "냉장고 파먹기"
def web_crawl(query):
  driver = webdriver.Chrome(options=chrome_options)
  query = query
  url = f"https://search.naver.com/search.naver?ssc=tab.kin.kqna&where=kin&sm=tab_jum&query={query}"
  driver.get(url) # 여기에 실제 URL을 입력하세요


  # 페이지 하단으로 스크롤 다운
  for i in range(30): #50 = 500개
    print(i)
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);") # 스크롤 다운 후 데이터 로딩을 위해 추가로 기다리기
    time.sleep(5) # 필요에 따라 시간 조절

  # 클래스 이름으로 요소 찾기
  questions = driver.find_elements(By.CLASS_NAME, 'api_txt_lines.question_text')

  # 찾은 요소의 텍스트 추출 및 출력
  question_list = [question.text for question in questions]
  df = pd.DataFrame({"questions":question_list})

  save_df = f'/content/drive/MyDrive/KITA_1026/project/data/questions/questions_{query}.csv'
  df.to_csv(save_df, encoding='utf-8-sig', index=False)

  # 브라우저 닫기
  driver.quit()

  return df
```
